# Testing model

```python
import os
from typing import List
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from src import prediction_module

from src import lib 
```

Loading data

```python
SAMPLE_SIZE = 20
test_data_dir = lib.DEFAULT_TRAINING_ARGS["test_data_dir"]
```

```python
# Sample 10 image paths from the test data with seed
np.random.seed(42)
image_names = np.random.choice(os.listdir(test_data_dir), SAMPLE_SIZE, replace=False)
image_paths = [f"{test_data_dir}/{image_name}" for image_name in image_names]
image_paths[0:3]
```

Predict

```python
# Make a prediction for every image in the test data
predictions = []
for image_path in image_paths:
    prediction = prediction_module.main(image_path=image_path, checkpoint_path=lib.BEST_CHECKPOINT_PATH, device='cpu')
    predictions.append(prediction)
```

Plot

```python
lib.plot_images_with_predictions(image_paths, predictions)
```
